```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.rcParams.update({'font.size': 14})
import seaborn as sns
```

```python
x_axis = [1, 2, 5, 10, 20, 40, 50, 60, 100]
sequential = [1.511, 2.763, 6.854, 14.972, 26.859, 54.046, 74.214, 99.436, 137.134]
fastflow = [1.683, 2.123, 3.891, 6.691, 12.783, 24.842, 30.669, 36.011, 60.689]
fastflow_nomap_blk = [0.837, 1.432, 3.466, 6.466, 11.958, 23.704, 29.667, 36.511, 61.003]
#fastflow_farm = [1.925, 2.082, 4.508, 7.013, 11.091, 21.937, 30.542, 32.854, 56.197]
fastflow_farm = [0.838, 1.404, 3.352, 6.339, 12.473, 24.289, 31.741, 36.659, 62.229]
threadpool = [0.602, 1.178, 3.329, 6.165, 12.091, 24.797, 29.884, 35.909, 63.134]
processpool = [0.922, 1.911, 5.790, 9.330, 19.205, 39.247, 48.665, 56.713, 92.970]
```

```python
fig, axs = plt.subplots(figsize=(18, 5), dpi=300)
colors = ["#fdd49e", "#fc8d59", "#cc4c02", "#6a51a3"]
linewidth = 2
style = 'o-'
plt.plot(x_axis, sequential, style, label="Sequential", linewidth=linewidth, color=colors[0])
plt.plot(x_axis, processpool, style, label="Processpool", linewidth=linewidth, color=colors[1])
plt.plot(x_axis, threadpool, style, label="Threadpool", linewidth=linewidth, color=colors[2])
plt.plot(x_axis, fastflow_nomap_blk, style, label="FastFlow", linewidth=linewidth, color=colors[3])

axs.set_xlabel("Number of tasks")
axs.set_ylabel("Time (seconds)")
axs.grid()
# Setting the number of ticks
axs.xaxis.set_major_locator(ticker.MultipleLocator(5))
#axs.yaxis.set_major_locator(ticker.MultipleLocator(2))
axs.yaxis.set_major_formatter(ticker.ScalarFormatter())
axs.yaxis.set_minor_formatter(ticker.ScalarFormatter())
axs.tick_params(which='minor', length=0)

plt.yscale("log")
plt.title('Time over number of tasks', fontweight="bold")
plt.legend()
plt.savefig("zerodivision", bbox_inches="tight")
plt.show()
```
